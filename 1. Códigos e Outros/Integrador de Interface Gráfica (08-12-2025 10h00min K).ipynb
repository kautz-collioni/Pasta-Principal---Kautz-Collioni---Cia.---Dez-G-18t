## **Integração à Interface Gráfica** ##

Converte os códigos desenvolvidos em Jupyter Notebook para o formato .py, garantindo compatibilidade funcional para posterior integração com a interface gráfica.

```python
with open("Script em Formatação (29-12-2025 16h59min K).ipynb", "r", encoding = "utf-8") as f:
    notebook = nbformat.read(f, as_version = 4)

code = ""
for cell in notebook.cells:
    if cell.cell_type == "code":
        code += cell.source + "\n\n"

code = code.rstrip("\n")

with open("../6. GitHub/Conector.py", "w", encoding = "utf-8") as f:
    f.write(code)
```
